## Decision Tree

#. Two types of algorithm can be used.  id3 and card algorith

#. leaf node is also known as outcome node.

#. for id3  -  entropy is used

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df= pd.read_csv("weather.csv")
```

```python
target = df['play']
```

```python
inputs = df.drop(['play'],axis=1)
```

```python
output=df.play
```

```python
le_temp  = LabelEncoder()
```

```python
le_humidity = LabelEncoder()
```

```python
le_windy = LabelEncoder()
```

```python
le_outlook = LabelEncoder()
```

```python
inputs['temp_n'] = le_temp.fit_transform(inputs['temp'])
```

```python
inputs['humidity_n'] = le_temp.fit_transform(inputs['humidity'])
```

```python
inputs['windy_n'] = le_temp.fit_transform(inputs['windy'])
```

```python
inputs['outlook_n'] = le_outlook.fit_transform(inputs['outlook'])
```

```python
inputs_n = inputs.drop(['outlook','temp','humidity','windy'],axis='columns')
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(inputs_n,target)
```

```python
model.predict([[2,1,1,1]])
```
